```python
N = str(5)
```

```python
filename = "pipeline/result"+N+"/result_out.100.topic_likelihoods"
with open(filename, "r") as f:
    topic_likelihoods = f.read()

topic_likelihoods = topic_likelihoods.split()
topic_likelihoods = map(lambda x: float(x), topic_likelihoods)
print topic_likelihoods
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(topic_likelihoods)
```

```python
filename = "pipeline/result"+N+"/result_out.100.topwords"
topwords = []
with open(filename, "r") as f:
    for line in f:
        line = line.split(':')
        topwords.append(line[0][:-1])

print topwords
```

```python
cuidict = "pipeline/result"+N+"/cui_dict.txt"
cui_dict = {}

with open(cuidict, 'r') as f:
    for line in f:
        l_split = line.split('\t')
        cui_dict[l_split[0]] = l_split[1][:-1]
```

```python
cui_words = []

for word in topwords:
    cui_words.append(cui_dict[word])
```

```python
import pandas as pd

likelihood_df = pd.DataFrame.from_dict({'cui': topwords, 'concept': cui_words, 'likelihood': topic_likelihoods})
print likelihood_df.head()
```

```python
likelihood_df_sorted = likelihood_df.sort('likelihood')
print likelihood_df_sorted
```

```python
likelihood_df_sorted.to_csv("pipeline/result6/result_out.100.topwords.likelihoods")
```
