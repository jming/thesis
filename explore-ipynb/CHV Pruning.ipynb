```python
import pandas as pd
```

```python
column_names = ['cui', 'term', 'chv_pref_name']
filename = "../pipeline/get_q/chv.tsv"
chv_df = pd.read_csv(filename, sep='\t', names=column_names, usecols=[0,1,2])
print chv_df.head()
```

```python
mrsty_df = pd.read_csv("../pipeline/get_q/mrsty.csv", names=['cui', 'category', 'category_name'], usecols=[0,1,2])
print mrsty_df.head()
```

```python
joined_df = chv_df.merge(mrsty_df, on='cui')
print joined_df.head()
```

```python
grouped_df = joined_df.groupby('category')
print grouped_df.head()
```

```python
groups = dict(list(grouped_df))
group_list = groups.keys()
print group_list
```

```python
categories = {}
for group in group_list:
    for category_name in groups[group]['category_name']:
        categories[group] = category_name
        #print category_name
        break
    #print group
    #break
for k in categories:
    print k, categories[k]
```

```python
from collections import OrderedDict

categories_ordered = OrderedDict(sorted(categories.items(), key=lambda t: t[0]))
for c in categories_ordered:
    print c, categories_ordered[c]
```

```python
with open("pipeline/get_q/mrsty_categories.txt", 'a') as f:
    for c in categories_ordered:
        f.write(c+','+categories_ordered[c]+'\n')
```
