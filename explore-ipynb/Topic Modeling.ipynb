```python
import random
import numpy as np
```

```python
"""

create a test post x word matrix with the following qualities:

* anchor posts are 1, 2, 3
    word 1 only appears in post 1,
    word 2 only appears in post 2,
    word 3 only appears in post 3,
    the rest of the words are random nonzero quantities

* the remaining posts are linear combinations of posts 1,2,3
    px = a1p1 + a2p2 + a3p3,
    where a1,a2,a3 are elements in {0,1,2}

"""

def create_example(num_words, num_posts):
    
    matrix = []
    
    # create the first three entries
    for i in range(3):
        post = np.concatenate((np.zeros(3),
            np.random.randint(1, high=10, size=num_words-3)))
        post[i] = 1
        matrix.append(post)
       
    # fill in remaining as linear combinations
    for i in range(4,num_posts):
        
        post = np.zeros(num_words)
        for j in range(3):
            post += float(random.randint(0,2)) * matrix[j]
        matrix.append(post)
    
    # return matrix
    return matrix
```
