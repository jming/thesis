```python
def read_matrix(filename):

    # return sparse.dok_matrix(sio.loadmat(filename))
    with open(filename, 'rb') as f:
        x = pickle.load(f)
    return x
```

```python
def calc_svd(matrix,k_in=6):

    U,s,Vh = scipy.sparse.linalg.svds(matrix,k=k_in)
    # V = Vh.T 

    print s
    # [  52.06867981   53.17331696   58.20323181   74.85311127  109.54414368 291.88696289]

    plt.plot(s)
    plt.show()
```

```python
print 'reading matrix', time.time()
matrix = read_matrix('cui_matrix.pickle')
print matrix
```

```python
ks = [6,10,15,20,50]

for k in ks:

    print 'calc svd k=',k, time.time()
    calc_svd(matrix,k)
```

```python
def calc_eigenvalues(matrix,k_in):
    w,v = scipy.sparse.linalg.eigs(matrix, k=k_in)
    
    print w
    plt.plot(w)
    plt.show()
    
    print v
    plt.hist(v, bins=25)
    plt.show()
```

```python
ks = [6, 10, 15]

for k in ks:
    print 'k=',k
    calc_eigenvalues(matrix,k)
```

```python
from sklearn import cluster
# from sklearn.preprocessing import StandardScaler

def spectral_clustering(matrix):
    
    spectral = cluster.SpectralClustering(n_clusters=10)
#     X = StandardScaler().fit_transform(matrix)
#     t0 = time.time()
    spectral.fit(matrix)
#     t1 = time.time()
    
    plt.scatter(matrix[:,0], matrix[:,1])
    plt.show()
```

```python
spectral_clustering(matrix)
```

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist, squareform
def ward_clustering(matrix):

    a = pdist(matrix)
    z = hierarchy.linkage(a, method='ward')
    
    d = hierarchy.dendrogram(z)
```

```python
ward_clustering(matrix)
```
